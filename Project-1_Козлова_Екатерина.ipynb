Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
df_copy_4 = df_copy_3.copy()

# Создание столбцов
df_copy_4['Город'] = None
df_copy_4['Готовность к переезду'] = None
df_copy_4['Готовность к командировкам'] = None

# Функция для классификации города
def classify_city(text):
    
    # Списки городов
    million_cities = ['Новосибирск', 'Екатеринбург','Нижний Новгород','Казань', 'Челябинск','Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж','Волгоград']
    big_cities = ['Москва', 'Санкт-Петербург']

    if not isinstance(text, str): # пропускаем если не строка
        return None
    city = text.split(',')[0].strip()
    
    if city in big_cities: # проверка Москвы и Санкт-Петергбурга
        return city
    elif city in million_cities: # проверка городов-миллиоников
        return 'город-миллионник'
    else:                       # проверка всех городов
        return 'другие'

# Функция для определения готовности к переезду
def check_relocate(text):
    
    # Фразы для определения готовности к переезду
    t_relocate = ['готов к переезду', 'готова к переезду', 'хочу переехать']
    f_relocate = ['не готова к переезду', 'не готов к переезду']
    
    # Проверяем отрицательные фразы
    for phrase in f_relocate:
        if phrase in text:
            return False
    
    # Проверяем положительные фразы
    for phrase in t_relocate:
        if phrase in text:
            return True
    
    # Если ничего не нашли
    return False

# Функция для определения готовности к командировкам
def check_b_trip(text):
    
    # Фразы для определения готовности к командировкам
    t_business_trip = ['готов к командировкам', 'готова к командировкам', 'готов к редким командировкам', 'готова к редким командировкам']
    f_business_trip = ['не готов к командировкам', 'не готова к командировкам']
    
    # Проверяем отрицательные фразы
    for phrase in f_business_trip:
        if phrase in text:
            return False
    
    # Проверяем положительные фразы
    for phrase in t_business_trip:
        if phrase in text:
            return True
    
    # Если ничего не нашли
    return False


# Применяем функции
df_copy_4['Город'] = df_copy_4['Город, переезд, командировки'].apply(classify_city)
df_copy_4['Готовность к переезду'] = df_copy_4['Город, переезд, командировки'].apply(check_relocate)
df_copy_4['Готовность к командировкам'] = df_copy_4['Город, переезд, командировки'].apply(check_b_trip)

# Удаление столбца
del df_copy_4['Город, переезд, командировки']

# Вывод
display(df_copy_4.head(3))
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>

Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
df_copy_5 = df_copy_4.copy()

# Функция типа занятости
def check_employment(text):
    
    # Создаем словарь для типов занятости
    employment_types = {
        'Полная занятость': ['полная занятость'],
        'Частичная занятость': ['частичная занятость'],
        'Проектная работа': ['проектная работа'],
        'Стажировка': ['стажировка'],
        'Волонтерство': ['волонтерство']
    }
    
    results = {}  # Словарь для результатов
    
    # Перебираем все типы занятости
    for col_name, keywords in employment_types.items():
        found = False 
        
        # Перебираем ключевые слова для текущего типа занятости
        for keyword in keywords:
            if keyword in str(text):  # Проверяем вхождение
                found = True
                break  # Прерываем цикл при первом совпадении
        
        results[col_name] = found  # Записываем результат
    
    return results

# Фукция опредения графика
def check_schedule(text):
    
    # Создаем словарь для типов графика
    schedule_types = {
        'Полный день': ['полный день'],
        'Вахтовый метод': ['вахтовый метод'],
        'Удаленная работа': ['удаленная работа'],
        'Гибкий график': ['гибкий график'],
        'Сменный график': ['сменный график']
    }
    
    results = {}  # Словарь для результатов

    for col_name, keywords in schedule_types.items():
        # Проверяем наличие любого из ключевых слов
        results[col_name] = any(keyword in str(text) for keyword in keywords)
    
    return results

# Создаем столбцы и заполняем их
employment_results = df_copy_5['Занятость'].apply(check_employment)
df_copy_5 = pd.concat([df_copy_5, employment_results.apply(pd.Series)], axis=1)

schedule_results = df_copy_5['График'].apply(check_schedule)
df_copy_5 = pd.concat([df_copy_5, schedule_results.apply(pd.Series)], axis=1)

# Удаление столбцов
del df_copy_5['Занятость']
del df_copy_5['График']

# Вывод
display(df_copy_5.head(3))
```

Вспомогательные инструменты для расчетов и оформления.

```python

df_result = df_6_merge.copy()

# Расчет статистик
class AgeStats:
    def __init__(self, data, col):
        self.мода = data[col].mode()[0]
        self.мин = data[col].min()
        self.макс = data[col].max()
        self.среднее = data[col].mean()
        self.медиана = data[col].median()
        self.стандартное_отклонение = data[col].std()
        self._25 = data[col].quantile(0.25)
        self._75 = data[col].quantile(0.75)
        self.стандартный_интервал = [
            self.среднее - self.стандартное_отклонение,
            self.среднее + self.стандартное_отклонение
        ]
        self.доверительный_интервал = [
            self.среднее - 2 * self.стандартное_отклонение,
            self.среднее + 2 * self.стандартное_отклонение
        ]
        self.высокий_доверительный_интервал = [
            self.среднее - 3 * self.стандартное_отклонение,
            self.среднее + 3 * self.стандартное_отклонение
        ]

# Вывод статистик
def stats(data, col):
    return AgeStats(data, col)

# Проверка аномалий
def anomaly_check (data, col):
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[col] < lower_bound) | (data[col] > upper_bound)]
    
    result = (
        f"Обнаружено {len(outliers)} потенциальных аномалий:\n"
        f"{outliers[[col]].describe()}"
    )
    return result

# Метод Тьюки
def outliers_iqr_mod(data, feature, left=1.5, right=1.5):
    if isinstance(data, str):
        data = pd.read_csv(data)
    x = data[feature]
    quartile_1, quartile_3 = x.quantile(0.25), x.quantile(0.75),
    iqr = quartile_3 - quartile_1
    lower_bound = quartile_1 - (iqr * left)
    upper_bound = quartile_3 + (iqr * right)
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x >= lower_bound) & (x <= upper_bound)]
    return outliers, cleaned

# Метод 3 сигм
def outliers_z_score_mod(data, feature, left=3, right=4, log_scale=False):
    if log_scale:
        x = np.log(data[feature] + 1)
    else:
        x = data[feature]
    
    mu = x.mean()
    sigma = x.std()
    
    lower_bound = mu - left * sigma
    upper_bound = mu + right * sigma
    
    outliers_2 = data[(x < lower_bound) | (x > upper_bound)]
    cleaned_2 = data[(x >= lower_bound) & (x <= upper_bound)]
    
    return outliers_2, cleaned_2
    
```

4. Постройте диаграмму, которая показывает зависимость **медианной** желаемой заработной платы (**"ЗП (руб)"**) от уровня образования (**"Образование"**). Используйте для диаграммы данные о резюме, где желаемая заработная плата меньше 1 млн рублей.
*Сделайте выводы по представленной диаграмме: для каких уровней образования наблюдаются наибольшие и наименьшие уровни желаемой заработной платы? Как вы считаете, важен ли признак уровня образования при прогнозировании заработной платы?*

```python
# 1. Предобработка данных
# Фильтруем данные
df_filtered = df_result[df_result['ЗП (руб)'] < 1_000_000]

# Уровни образования в нужном порядке
edu_order = ['среднее', 'среднее специальное', 'неоконченное высшее', 'высшее']

# 2. Визуализация
plt.figure(figsize=(8, 8))
ax = sns.boxplot(data=df_filtered, 
                x='Образование',
                y='ЗП (руб)', 
                hue='Образование',
                showfliers=False,
                palette='viridis',
                order=edu_order)

plt.title('Распределение желаемой зарплаты по уровню образования', y=1.01, fontsize=18)
plt.xlabel('Уровень образования', labelpad=10, fontsize=14)
plt.ylabel('Желаемая ЗП, руб',  labelpad=10, fontsize=14)
plt.yticks(np.arange(0, 200000, 10000)) # Устанавливаем более частые деления на оси Y 
plt.xticks
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()

plt.show()

# 3. Таблица
stats_list = []

for edu_level in edu_order:
    group = df_filtered[df_filtered['Образование'] == edu_level]['ЗП (руб)'].dropna()
    
    q1 = group.quantile(0.25)
    median = group.quantile(0.50)
    q3 = group.quantile(0.75)
    iqr = q3 - q1
    
    lower_whisker = group[group >= (q1 - 1.5 * iqr)].min()
    upper_whisker = group[group <= (q3 + 1.5 * iqr)].max()
    
    stats_list.append({
        'Образование': edu_level,
        'Верхний ус': int(upper_whisker),
        'Q3 (75%)': int(q3.round(0)),
        'Медиана (50%)': int(median),
        'Q1 (25%)': int(q1.round(0)),
        'Нижний ус': int(lower_whisker)
    })

df_stats = pd.DataFrame(stats_list)

display(df_stats.T)
```

**Вывод:**

Ниже всех оказались зарплаты у соискателей со средним и средним специальным образованием. Медианная зарплата 40000 рублей, 25 квантиль 30000 рублей. Но 75 квантиль у людей со средним специальным образованием ниже, чем у людей со средним - 50000 рублей против 60000 рублей. Больше зарплатные ожидания у специалистов с неоконченным высшим образованием: медиана - 50000 рублей, 25% - 32750 рублей, 75% - 70000 рублей. У специалистов с высшим образованием самые высокие зарплаты: медиана - 60000 рублей, 25% - 40000 рублей, 75% - 100000 рублей. При этом у специалистов с высшим образованием самый большой разброс (75%-25%), при этом максимальная зарплата существенно выше, чем у всех остальных категорий. Для высокой зарплаты предпочтительно иметь высшее образование.
